```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time) 
```

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
urlFotografifirmor = "https://kulturnav.org/api/list/deb494a0-5457-4e5f-ae9b-e1826e0de681?lang=sv " 
urlFotografer = "https://kulturnav.org/api/list/1f368832-7649-4386-97b6-ae40cce8752b?lang=sv " 
r = http.request('GET', urlFotografifirmor)
try:
    data = json.loads(r.data)
    dfFirma = pd.json_normalize(data)
except Exception as e: print(e)
```

```python
dfFirma
```

```python
dfFirma[{"caption.sv","uuid"}]
```

```python
dfFirma[{"caption.sv","uuid"}].to_csv("kulturnav_fotografifirmor.csv",index=False)
```
